```python
from skimage.feature import hog
import matplotlib.pyplot as plt

# Load an image from file
I = plt.imread('./assets/person.png')

fd, hogI = hog(I, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=True, channel_axis=-1)
# Show original image and HOG image in subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(I, cmap='gray')
ax[0].set_title('Own implementation of HOG Image')
ax[0].axis('off')
ax[1].imshow(hogI, cmap='gray')
ax[1].set_title('HOG Image')
ax[1].axis('off')
plt.show()
```
